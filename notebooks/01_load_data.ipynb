# Load data

```python
#importanbdo bibliotecas
import os

import yfinance as yf
```

```python
os.makedirs('data/raw', exist_ok=True)


ticker = 'AAPL'
df = yf.download(ticker, start='2018-01-01')
df = df.droplevel(1, axis=1).rename_axis(None, axis=1)
df = df.reset_index()
df.head()
```

```python
df.to_pickle('data/raw/raw_data.pkl')
```
