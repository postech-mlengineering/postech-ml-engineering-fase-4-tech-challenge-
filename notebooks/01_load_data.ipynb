## 1. Setup para extração dos dados

Nesta etapa, importamos as bibliotecas necessárias para realizar a extração dos dados históricos de mercado. Usamos a biblioteca `yfinance` para baixar os dados do Yahoo Finance e `pathlib` para gerenciar os caminhos dos arquivos e diretórios de forma robusta e independente de sistema operacional.

```python
from pathlib import Path

import yfinance as yf
```

## 2. Configuracoes

Definimos os parâmetros básicos para a extração dos dados (como o ticker do ativo e a data inicial de busca) e criamos a estrutura de diretórios necessária para armazenar os dados brutos extraídos.

```python
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
DATA_DIR = PROJECT_ROOT / "data"
RAW_DATA_DIR = DATA_DIR / "raw"

TICKER = "AAPL"
START_DATE = "2018-01-01"

RAW_DATA_DIR.mkdir(parents=True, exist_ok=True)
```

## 3. Coleta de dados

Fazemos o download dos dados históricos do ticker especificado a partir do Yahoo Finance e os salvamos localmente em formato CSV na pasta de dados brutos (`raw`).

```python
raw_data = yf.download(TICKER, start=START_DATE, progress=False, auto_adjust=False)

if raw_data.empty:
    raise ValueError(f"Nenhum dado foi retornado para o ticker {TICKER}.")

raw_data = raw_data.reset_index()
raw_path = RAW_DATA_DIR / f"{TICKER}.csv"
raw_data.to_csv(raw_path, index=False)

raw_data.head()
```
